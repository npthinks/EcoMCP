```python
import pandas as pd

def clean_region(value):
    if pd.isna(value):
        return None
    # Convert to string and split by comma
    parts = [p.strip().capitalize() for p in str(value).split(",") if p.strip()]
    # Remove duplicates and sort alphabetically for consistency
    parts = sorted(set(parts))
    # Join back with consistent formatting
    return ", ".join(parts)

df["Region"] = df["Region"].apply(clean_region)
```

```python
df["Region"] = df["Region"].astype("str").str.strip()

def clean_region(value):
    if pd.isna(value):
        return None
    # Convert to string and split by comma
    parts = [p.strip().capitalize() for p in str(value).split(",") if p.strip()]
    # Remove duplicates and sort alphabetically for consistency
    parts = sorted(set(parts))
    # Join back with consistent formatting
    return ", ".join(parts)

df["Region"] = df["Region"].apply(clean_region)
```

```python
df["ID"] = df["Region"].map(id_dict)
```

```python
df["ID"] = df.groupby("Region").cumcount() + 1
df["ID"] = df.apply(lambda row: id_dict[row["Region"]] + row["ID"], axis=1)
```

```python
df.set_index("ID", inplace = True)
```

```python
df.reset_index(drop=False, inplace=True)
```
